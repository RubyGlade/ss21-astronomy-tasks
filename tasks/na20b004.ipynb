```python
#for task 4
primary = hdulist[0]
l = []
l.append(primary.header['NAXIS3'])
l.append(primary.header['CTYPE3'])
l.append(primary.header['CUNIT3'])
l.append(primary.header['TELESCOP'])
l.append(primary.header['OBSERVER'])
print(l)
```

```python
#for task 6
font1 = {'family':'serif', 'size':40}
font2 = {'family':'serif', 'size':25}
font3 = {'family':'serif', 'size':20}
font4 = {'family':'serif', 'size':15}
plt.figure(1, (96, 10))
plt.title('TASK 6', fontdict = font1)
plt.imshow(first_slice, cmap = "gray")
plt.xlabel("x - axis", fontdict = font2)
plt.ylabel('y - axis', fontdict = font2)
plt.colorbar()
```

```python
#for task 7
i = 0
while(i<30):
    if(i==0):
        stack = data[0]
    else:
        stack = stack + data[i]
    i = i + 5
print("Maximum pixel: ", stack.max())
print("Minimum pixel: ", stack.min())
print("Standard deviation of all the pixels: ", stack.std())
plt.figure(1, (96, 10))
plt.title('TASK 7', fontdict = font1)
plt.imshow(stack, cmap = "gray")
plt.xlabel("x - axis", fontdict = font2)
plt.ylabel('y - axis', fontdict = font2)
plt.colorbar()
```

```python
#for task 8
def function_to_plot_altitude_vs_time(coordinates, time):
    coordinates.transform_to('galactic')
    iitm = EarthLocation(lat=12.992*u.deg, lon=80.234*u.deg, height=50*u.m)
    time_actual = time - 5.5*u.hour
    t = np.array([])
    for i in range(1, 100):
        t = np.append(t, (i * 12.0)/100)
    time_array = (t * u.hour) + time_actual
    coords_h = coordinates.transform_to(AltAz(obstime = time_array, location = iitm))
    plt.plot(t, coords_h.alt)
    plt.title('TASK 8', fontdict = font3)
    plt.xlabel("Time since mentioned time(hours)", fontdict = font4)
    plt.ylabel('Altitude(degrees)', fontdict = font4)
    return plt.show()
```
